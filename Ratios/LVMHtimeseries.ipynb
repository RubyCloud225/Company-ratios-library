```python
from time_series import StockPricePredictor
from datetime import datetime, timedelta
```

```python
predictor = StockPricePredictor("MC.PA", "2022-12-31", "2024-06-30")
predictor.decompose_time_series()
predictor.fit_arma_model()
predictor.fit_sarima_model()
```

```python
y_pred_df = predictor.fit_arma_model()
last_pred = y_pred_df.iloc[-1]
```

```python
current_date = datetime.strptime(predictor.end_date, '%Y-%m-%d')
next_day = current_date + timedelta(days=1)
```

```python
print(f'Predicted Share price for {next_day.strftime("%Y-%m-%d")}: {last_pred.values[0]:.2f}' )
```
